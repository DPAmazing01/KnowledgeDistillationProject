```python
from vgg16 import *
from resnet18 import *
from densenet import *
from train import *
from summary import *
import torch
import numpy as np
import time

device = 'cuda' if torch.cuda.is_available() else 'cpu'
```

### Teacher & student model training

```python
VGG16_teacher_net = VGG16()
VGG16_teacher_net = VGG16_teacher_net.to(device)
VGG16_student_net = VGG16_half()
VGG16_student_net = VGG16_student_net.to(device)
ResNet18_teacher_net = ResNet18()
ResNet18_teacher_net = ResNet18_teacher_net.to(device)
ResNet18_student_net = ResNet18_half()
ResNet18_student_net = ResNet18_student_net.to(device)
DenseNet_teacher_net = DenseNet()
DenseNet_teacher_net = DenseNet_teacher_net.to(device)
DenseNet_student_net = DenseNet_half()
DenseNet_student_net = DenseNet_student_net.to(device)
```

```python
start_time = time.time()
train(VGG16_teacher_net, epochs=100, batch_size=128, lr=0.01, reg=5e-4)
end_time = time.time()
elapsed_time = end_time - start_time
```

```python
VGG16_teacher_net.load_state_dict(torch.load("VGG16_teacher_net.pt"))
print("-----VGG16 Teacher Net-----")
summary(VGG16_teacher_net)
test(VGG16_teacher_net)
minutes, seconds = divmod(elapsed_time, 60)
print(f"VGG16 Teacher Net Training Time: {int(minutes)} minutes and {int(seconds)} seconds")
print("-------------------------------")
```
